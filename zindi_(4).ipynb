```python
#FIXING OUTLIERS
#log transformer
def log_transformer(data):
    data = np.log(data) + 1
    return data
    
        
def winsorizer(data,x,lower,upper):
    per_10th = data[x].quantile(lower)
    per_90th = data[x].quantile(upper)
    data[x] = np.where(data[x] < per_10th, per_10th, data[x])
    data[x] = np.where(data[x] > per_90th, per_90th, data[x])
    return data[x]


def imputation(data, metrics, lim = 3):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q1 - Q3
    data = np.where(data < Q1 - lim * IQR, metrics, data)
    data = np.where(data > Q3 - lim * IQR, metrics, data)
    return data
```

```python
data[outliers] = winsorizer(data, outliers, 0.1, 0.9)
data[outliers]
```

```python

# CONSTANT REPLACEMENT
def nan_replacer(X, method):
    if method.lower() == "mean":
        X.iloc[:,:] =X.iloc[:,:].fillna(X.iloc[:,:].mean())
        return X.iloc[:,:]
    elif method.lower()== "median":
        X.iloc[:, 4:] = X.iloc[:, 4:].fillna(X.iloc[:, 4:].median())
        return X.iloc[:, 4:]
    elif method.lower() == "mode":
        X.iloc[:, 4:] = X.iloc[:, 4:].fillna(X.iloc[:, 4:].mode()[0])
        return X.iloc[:, 4:]
    elif method.lower() == "zero":
        X.iloc[:, 4:] = X.iloc[:, 4:].fillna(0)
        return X.iloc[:, 4:]
        
#NON-RANDOM REPLACMENT
def nan_non_ran_replacer(X, method):
    if method.lower() == "ffill":
        X.iloc[:, 4:] = X.iloc[:, 4:].fillna(method = "ffill")
        return X.iloc[:, 4:]
    elif method.lower() == "bfill":
        X.iloc[:, 4:] = X.iloc[:, 4:].fillna(method = "backfill")
        return X.iloc[:, 4:]
    elif method.lower() == "knn":
        from sklearn.impute import KNNImputer
        knn_imputer = KNNImputer(n_neighbors = 7)
        X.iloc[:, 4:] = knn_imputer.fit_transform(X.iloc[:, 4:])
        return X.iloc[:, 4:]
        

    
# RANDOM REPLACEMENT
```

```python
##Scaling fuction
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler, PowerTransformer
def Scaler (data, col, method):
    scaled_data = data[col]
    if method.lower() == "mmax":
        scaler = MinMaxScaler()
        scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data))
        scaled_data.columns = col
        return scaled_data
    elif method.lower() == "sts":
        scaler = StandardScaler()
        scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data))
        scaled_data.columns = col
        return scaled_data
    elif method.lower() == "rs":
        scaler = RobustScaler()
        scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data))
        scaled_data.columns = col
        return scaled_data
    elif method.lower() == "ptb":
        scaler = PowerTransformer(method = "bos-cox")
        scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data))
        scaled_data.columns = col
        return scaled_data
    elif method.lower() == "pty":
        scaler = PowerTransformer(method = "yeo-johnson")
        scaled_data = pd.DataFrame(scaler.fit_transform(scaled_data))
        scaled_data.columns = col
        return scaled_data
```
